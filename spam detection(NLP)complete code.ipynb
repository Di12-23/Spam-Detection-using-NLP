Text Preprocessing

```python
# Convert text to lowercase
messages['message'] = messages['message'].str.lower()

messages.head(5)
```

```python
import string
```

```python
mess = 'Sample message! Notice that : it has punctuation     .'

# Check characters to see if they are in punctuation
nopunc = [char for char in mess if char not in string.punctuation]

# Join the characters again to form the string after removing all punctuation.
nopunc = ''.join(nopunc)
```

```python
nopunc.split()
```

```python
# Now just remove any stopwords
clean_mess = [word for word in nopunc.split() if word not in stopwords.words('english')]
```

```python
clean_mess
```

```python
def text_process(mess):
    """
    Takes in a string of text, then performs the following:
    1. Remove all punctuation
    2. Remove all stopwords
    3. Returns a list of the cleaned text
    """
    # Check characters to see if they are in punctuation
    nopunc = [char for char in mess if char not in string.punctuation]

    # Join the characters again to form the string.
    nopunc = ''.join(nopunc)
    
    # Now just remove any stopwords
    return [word for word in nopunc.split() if word not in stopwords.words('english')]
```

```python
messages.head(5)
```

```python
# Check to make sure its working
messages['message'].head(5).apply(text_process)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Initialize and fit CountVectorizer
bow_transformer = CountVectorizer(analyzer=text_process).fit(messages['message'])
```

```python
# Print total number of vocab words
print(f'Total vocabulary words: {len(bow_transformer.vocabulary_)}')
```

```python
# Take one text message and get its bag-of-words counts as a vector
message4 = messages['message'][3]
print(f'Message 4: {message4}')
```

```python
# Vector representation of the message
bow4 = bow_transformer.transform([message4])
print(f'Vector representation of message 4: {bow4}')
print(f'Shape of vector: {bow4.shape}')
```

```python
# Checking words that appear twice
indices_with_counts = bow4.nonzero()[1]
counts = bow4.data
word_counts = dict(zip(indices_with_counts, counts))
```

```python
# Print words that appear twice
for idx, count in word_counts.items():
    if count == 2:
        word = bow_transformer.get_feature_names_out()[idx]
        print(f"Word appearing twice: {word} at index {idx}")
```

```python
# Transform the entire DataFrame of messages
messages_bow = bow_transformer.transform(messages['message'])
```

```python
# Print shape of the sparse matrix and number of non-zero occurrences
print(f'Shape of Sparse Matrix: {messages_bow.shape}')
print(f'Amount of Non-Zero occurrences: {messages_bow.nnz}')
```

```python
# Calculate and print sparsity
sparsity = (100.0 * messages_bow.nnz / (messages_bow.shape[0] * messages_bow.shape[1]))
print(f'Sparsity: {round(sparsity, 2)}%')
```

Train Test Split

```python
from sklearn.model_selection import train_test_split

msg_train, msg_test, label_train, label_test = \
train_test_split(messages['message'], messages['label'], test_size=0.2)

print(len(msg_train), len(msg_test), len(msg_train) + len(msg_test))
```

```python
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ('bow', CountVectorizer(analyzer=text_process)),  # strings to token integer counts
    ('tfidf', TfidfTransformer()),  # integer counts to weighted TF-IDF scores
    ('classifier', MultinomialNB()),  # train on TF-IDF vectors w/ Naive Bayes classifier
])
```

```python
pipeline.fit(msg_train,label_train)
```

```python
predictions = pipeline.predict(msg_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(predictions,label_test))
```
